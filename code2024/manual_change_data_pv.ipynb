```python
import os
import pandas as pd
import plotly.graph_objs as go

def read_and_plot(year,month,day):
    month_str = f"{month:02d}"
    day_str = f"{day:02d}"
    
    P_df = pd.read_csv(f'../preprocessed_data/data_PV_1minresample_daily/PV-{year}-{month_str}/pv_8kW_1minresample_{year}{month_str}{day_str}.csv', header=0, parse_dates=['datetime'], index_col='datetime')
    I_df = pd.read_csv(f'../preprocessed_data/data_Irradiance_1minresample_daily/I-{year}-{month_str}/irradiance_1minresample_{year}{month_str}{day_str}.csv', header=0, parse_dates=['Datetime'], index_col='Datetime')
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=P_df.index, y=P_df['Ptot (kW)'], mode='lines', name='Power', yaxis='y1'))
    fig.add_trace(go.Scatter(x=I_df.index, y=I_df['Irradiance (W/m2)'], mode='lines', name='Irradiance', yaxis='y2'))
    fig.update_layout(
        title='Power and Irradiance vs. Time',
        xaxis_title='Datetime',
        yaxis=dict(
            title='Power (kW)',
            side='left'
        ),
        yaxis2=dict(
            title='Irradiance (W/m2)',
            overlaying='y',
            side='right'
        ),
        legend_title='Metrics',
    )
    fig.show()

    return P_df, I_df

    
```

# 2024-10-28 data power and irradiance not match

```python
y,m,d = 2024,10,28
P_df, I_df = read_and_plot(y,m,d)
```
